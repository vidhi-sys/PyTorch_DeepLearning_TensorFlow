```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python

target = torch.tensor([2.5, 5.0, 4.0, 8.0])
preds = torch.tensor([3.0, 5.0, 2.5, 7.0])
```

```python
def MSE(x,y):
  loss=target-preds
  loss_sqr=loss*loss
  avg_loss=loss_sqr.mean()
  return avg_loss
```

```python
print(MSE(target,preds))
```

```python
plt.plot(target.numpy(), label='target')
plt.plot(preds.numpy(), label='preds')
plt.legend()
plt.show()
```

```python
preds_range_np = np.arange(0, 20.1, 0.1)
preds_range = torch.tensor(preds_range_np, dtype=torch.float32)
```

```python
mse_values = []
target = torch.tensor([10.0])

for pred in preds_range:
  mse = MSE(target, pred.unsqueeze(0))
  mse_values.append(mse)

mse_values = torch.tensor(mse_values)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(preds_range.numpy(), mse_values.numpy())
plt.xlabel("Predictions")
plt.ylabel("MSE")
plt.title("MSE Curve for Target = 3")
plt.grid(True)
plt.show()
```
